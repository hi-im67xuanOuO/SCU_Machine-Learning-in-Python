```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples =150, #隨機產生150個點
                 n_features = 2, #
                 centers = 3, #密度很高的3群
                 cluster_std = 0.5,
                 random_state = 0)
```

```python
Complete = linkage(X, method = "complete", metric = "euclidean")
dendrogram(Complete)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

```python
from scipy.cluster.hierarchy import ward
from scipy.cluster.hierarchy import dendrogram

linkage = ward(X)
dendrogram(linkage)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(linkage = "complete",
                            affinity = "euclidean",
                            n_clusters = 3
y_km = ac.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0], #先選出y_km為0的index, 後面的0為該點的x座標、1為y座標
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')

plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')

plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')


plt.legend(scatterpoints = 1)
plt.grid()
plt.show()
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise = 0.05, random_state = 0)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
#K-Means++
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 2,
           init = 'k-means++',
           n_init = 10,
           max_iter = 300,
           tol = 1e-04,
           random_state = 0)

y_km = km.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen', edgecolor = 'black',
            label = 'cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange', edgecolor = 'black',
            label = 'cluster 2')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(linkage = "complete",
                            affinity = "euclidean",
                            n_clusters = 2)
y_km = ac.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen', edgecolor = 'black',
            label = 'cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange', edgecolor = 'black',
            label = 'cluster 2')
plt.show()
```
