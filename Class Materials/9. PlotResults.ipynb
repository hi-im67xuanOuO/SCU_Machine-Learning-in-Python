```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples = 150, #隨機生成150點
                 n_features = 2,
                 centers = 3, #分三群
                 cluster_std = 0.5, #密度0.5
                 random_state = 0)

import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1],
            c = 'white', marker = 'o', edgecolor = 'black', s = 50)
plt.grid()
plt.show()
```

```python
# K-means 

from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3,
            init = 'random',
            n_init = 10, # 以獨立不同、隨機選取的質心，執行演算法10次，並以最低SSE的模型作為最終模型
            max_iter = 300, #指定每次執行的最大迭代次數
            tol = 1e-04,
            random_state = 0)

y_km = km.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s = 250, marker = '*',
            c = 'red', edgecolor = 'black',
            label = 'centroids')

plt.legend(scatterpoints = 1)
plt.grid()
plt.show()
```

```python
## KMeans++

from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3,
            init = "k-means++",
            n_init = 30,
            max_iter = 300,
            tol = 1e-04,
            random_state = 0)

y_km = km.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s = 250, marker = '*',
            c = 'red', edgecolor = 'black',
            label = 'centroids')

plt.legend(scatterpoints = 1)
plt.grid()
plt.show()
```
